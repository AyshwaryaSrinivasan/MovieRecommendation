Importing required libraries

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

Extracting imdb links from links file

```python
linksData = pd.read_csv('Data/links.csv',usecols=[0,1])
movieIds = linksData['movieId'].tolist()
imdbIds = linksData['imdbId'].tolist()
```

```python
directors=[]
actors=[]
```

```python
for imdbId in imdbIds:
    url = 'http://www.imdb.com/title/tt'+ '0'*(7-len(str(imdbId))) + str(imdbId)+ '/'
    response = requests.get(url)
    print(url )
    if(response.status_code == 200):
        bs_movie = BeautifulSoup(response.text,'html.parser')
        director=bs_movie.find('div',class_='credit_summary_item').find_all('a')[0].text
        directors.append(director)
        actorsString = ""
        actorsRs = bs_movie.find_all('div',class_='credit_summary_item')[-1].find_all('a')
        for ac in actorsRs:
            temp=str(ac)[str(ac).index('>')+1:len(str(ac))]
            actorsString = actorsString+ temp[0:str(temp).index('<')]+'|' 
        actors.append(actorsString)
        
        
   
```

```python
df = pd.DataFrame(list(zip(movieIds,imdbIds, directors, actors)), 
               columns =['MovieId', 'ImdbId', 'Director','Cast']) 
df 
```

```python
df.to_csv ('Data/MovieDetails.csv', index = False, header=True)
```
